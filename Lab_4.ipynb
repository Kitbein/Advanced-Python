```python
from PIL import Image, ImageDraw, ImageFont
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
def createImg(text):
    img = Image.new('RGB',(1300, 500), color='white')
    draw = ImageDraw.Draw(img)
    w, h = 1000, 500
    draw.ellipse((100, 125, 250, 254), fill=(243,56,87))
    draw.rectangle((540, 100, 750, 300), fill=(63,234,13))
    draw.rectangle((150, 175, 400, 425), fill=(65,5,140))
    draw.ellipse((650, 200, 850, 400), fill=(243,56,87))
    text_font = ImageFont.truetype("arial.ttf", size = 40)
    width, height = draw.textsize(text, font=text_font)
    draw.text(((w-width)/2, (h-height)/2), text, font=text_font, fill=(0,0,0))
    return img

ukr = createImg("  ,  ,    !\n      ?\n       ,  ?\n     ,   ?")
eng = createImg(" Oh my dear thoughts, oh my dear thoughts,\n troubling are you to me!\n Why have you appeared on a paper in oh so sad rows?")
display(ukr)
```

```python
import os

def read_text(img, lng):
    if lng == "ukr":
        os.environ['TESSDATA_PREFIX'] = 'C:/Program Files/Tesseract-OCR/tessdata'
        print(pytesseract.image_to_string(img, lang = 'ukr'))
    else:
        print(pytesseract.image_to_string(img, lang = 'eng'))

display(eng)
read_text(eng, "eng")
```

```python
display(ukr)
read_text(ukr, "ukr")
```

```python

def binarize(image_to_transform, threshold):
    output_image=image_to_transform.convert("L")
    for x in range(output_image.width):
        for y in range(output_image.height):
            if output_image.getpixel((x,y))< threshold:
                output_image.putpixel( (x,y), 0 )
            else:
                output_image.putpixel( (x,y), 255 )
    return output_image

display(binarize(ukr, 38))
read_text(ukr, "ukr")
```
